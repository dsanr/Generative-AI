# Layout PDF Reader

```python
from llmsherpa.readers import LayoutPDFReader
```

```python
llmsherpa_api_url = "https://readers.llmsherpa.com/api/document/developer/parseDocument?renderFormat=all"
pdf_reader = LayoutPDFReader(llmsherpa_api_url)
```

```python
doc = pdf_reader.read_pdf('./pdf/how-to-formulate-successful-business-strategy.pdf')
```

```python
len(doc.chunks())
```

```python
len(doc.sections())
```

```python
len(doc.tables())
```

```python
for i in doc.chunks():
    print(i.to_context_text())
    # print(i.sentences)
    # print(i.sections())
    print('\n')
```

```python
for section in doc.sections():
    print(f'title - {section.title}')
    print(f'text - {section.to_context_text()}')
    print(section.to_text(include_children=True, recurse=True))
    print('\n')
```
